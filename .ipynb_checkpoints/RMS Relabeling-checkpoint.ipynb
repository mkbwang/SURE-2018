# RMS Relabeling
Before I carry out any data science, I want to relabel the numeric RMS to be "in control" or "not in control". This is because different people have different ability of performing the tracking task. We are not interested in the "absolute" performance. Instead we are interested in the person's internal psychological state(whether they are calm or not).

Below is the participants that took the binary test.

```python
participants = [0,2,4,7,9,10,13,15,17,20,22,24,27,29,31,34,36,38,41,43,45,48,50,52,55,57,59,62,
                64,66,69,71,72,74,76,77,79,82,86,88]
```

Have lineplots of all the RMSs for all the participants.

```python
sns.set()
pp = PdfPages('RMSplot.pdf')
for participant in participants:
    result = book.parse(names[3*participant])
    result = result.iloc[:,[4,6,8,12,15,16,17]]
    colnames = ['True_state','Alert','Identification','First_toggle_time','RMS','Attention','Click']
    result.columns = colnames
    xlab = 'Participant '+str(participant)
    plt.figure()
    _ = plt.plot(result['RMS'])
    _ = plt.xlabel(xlab)
    _ = plt.ylabel('RMS')
    plt.savefig(pp,format='pdf')
pp.close()
```

Plot the Inertia of Kmeans clustering with different number of clusters.

```python
ks = range(2, 10)
pp = PdfPages('Kmeans_Inertia.pdf')
for participant in participants:
    result = book.parse(names[3*participant])
    result = result.iloc[:,[4,6,8,12,15,16,17]]
    colnames = ['True_state','Alert','Identification','First_toggle_time','RMS','Attention','Click']
    result.columns = colnames
    result = result.drop(result['RMS'].idxmax())
    xlab = 'Participant '+str(participant)
    plt.figure()
    inertias = []
    for k in ks:
        # Create a KMeans instance with k clusters: model
        model = KMeans(n_clusters=k)
    
        # Fit model to samples
        model.fit(result['RMS'].reshape(-1,1))
    
        # Append the inertia to the list of inertias
        inertias.append(model.inertia_)

    # Plot ks vs inertias
    _ = plt.plot(ks, inertias, '-o')
    _ = plt.xlabel('Number of Clusters')
    _ = plt.ylabel('Inertia')
    _ = plt.title(xlab)
    _ = plt.xticks(ks)
    plt.savefig(pp,format='pdf')
pp.close()
```

Plot the histograms of RMSs for all the participants.

```python
sns.set()
pp = PdfPages('Histograms.pdf')
for participant in participants:
    sheet = book.parse(names[3*participant])
    plt.figure()
    _ = sheet['RMS'].hist(bins=15)
    _ = plt.xlabel('RMS')
    _ = plt.ylabel('Frequency')
    _ = plt.title('Participant '+str(participant))
    plt.savefig(pp,format='pdf')
pp.close()
```

Combine data of all the participants into one dataframe.

```python
def mapping(alert):
    if alert=='DANGER':
        return True
    else:
        return False
```

```python
def censor(toggle): 
    if toggle == -1:
        return 10000
    else:
        return toggle
```

```python
fulldata['Alert'] = fulldata['Alert'].apply(mapping)
```
